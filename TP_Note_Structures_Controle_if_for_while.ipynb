Écrire une fonction qui reçoit deux nombres `a` et `b` et retourne :
- le plus grand des deux
- ou la chaîne `'Égaux'` s’ils sont égaux



```python
def fonction_egaux(x,y):
    if x==y:
        print(f"Égaux")
    elif x>y:
        return x
    else: return y

a=float(input(f"Entrez le nombre 1"))
b=float(input(f"Entrez le nombre 2"))
fonction_egaux(a,b)
```

2) `'Multiple de 3'` ou `'Pas multiple de 3'`



```python
def fonction_pair_3(x):
    resultat=[]
    if x % 2==0:
     resultat.append("pair")
    else :
     resultat.append("impair")
    if x % 3==0:
     resultat.append("multiple de 3")
    else :
     resultat.append("pas multiple de 3")
    return resultat
a=int(input(f"Entrez le nombre"))
fonction_pair_3(a)
```

## Exercice 9 (3 pts)

Écrire un programme qui calcule et retourne la somme des nombres de 1 à 100 (inclus) en utilisant une boucle `for`.



```python
def somme(x, y):
    total = 0
    for i in range(x, y + 1):
        total = total + i
    return total    
resultat = somme(1,100)
print(f"Somme = {resultat}")
```

## Exercice 10 (2 pts)

Écrire une fonction qui reçoit une chaîne `mot` et retourne la liste de ses caractères (un caractère par élément) en utilisant une boucle `for`.


```python
def fonction_caractere(mot):
    liste=[]
    for c in mot:
        liste.append(c)
    return liste
mot=str(input("Entrez le mot"))
resultat = fonction_caractere(mot)
print(resultat)
```

## Exercice 12 (2 pts)

Écrire une fonction qui reçoit une liste de nombres saisis successivement et retourne la somme jusqu’à rencontrer un 0.

Exemple : `[5, 2, -1, 0, 100]` -> somme = `6`.

Consigne : utiliser une boucle `while`.




```python
def fonction_somme(liste):
    somme = 0
    i = 0
    while i < len(liste) and liste[i] != 0:
        somme += liste[i]
        i += 1
    return somme

resultat = fonction_somme([1, 2, 3, 4, 5, 0, 6])
print(resultat)
```
